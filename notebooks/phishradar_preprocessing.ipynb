```python
import pandas as pd
import numpy as np
```

```python
# load datasets
text_data = pd.read_csv("../dataset/phishing_text.csv")
url_data = pd.read_csv("../dataset/phishing_urls.csv")

text_data.head()
```

```python
text_data = text_data.drop(columns=["Unnamed: 0"])
```

```python
text_data = text_data.rename(columns={
    "Email Text": "text",
    "Email Type": "label"
})

text_data.head()
```

```python
text_data["label"] = text_data["label"].map({
    "Phishing Email": 1,
    "Safe Email": 0
})

text_data.head()
```

```python
text_data = text_data.dropna()

text_data.isnull().sum()

```

```python
text_data = text_data.drop_duplicates()

print(text_data.shape)
```

```python
import re

def clean_text(text):

    text = text.lower()

    text = re.sub(r"http\S+", "", text)

    text = re.sub(r"[^a-zA-Z ]", "", text)

    text = re.sub(r"\s+", " ", text)

    return text
```

```python
text_data["clean_text"] = text_data["text"].apply(clean_text)

text_data.head()
```

```python
url_data = url_data.drop(columns=[
    "FILENAME",
    "URL",
    "Domain",
    "Title",
    "TLD"
])

url_data.head()
```
